## The X Or probelm

can be solved using two neurons with one hidden layer. 

why:
* let's both neurons in the first layer can form a decision boundry
* the hidden layer neuron can modify the functions from the existing neurons i.e. map the data into a new space 
* the hidden layer makes prediction based on this modification




## Network with a single hidden layer

* input $\mathbf{x} \in \mathbb{R}^n$
* hidden $ \mathbf{W_1} \in \mathbb{R}^{m*n}, \mathbf{b_1} \in \mathbb{R}^m$
* output $\mathbf{W_2} \in \mathbb{R}^{o*m}, \mathbf{b_2} \in \mathbb{R}^o$
* activation function: $\sigma$
* hyperparameter: number of layers

*We have a network with 10 inputs, a first hidden layer with 50
neurons, a second hidden layer with 20 neurons and 5 output
neurons. How many trainable parameters does this network have?*

((current layer neurons c * previous layer neurons p)+1*c)

* input 10 neurons 
* hidden 
    * 50 neurons
    * 20 neurons
* output 5 neurons

$h_i=\sigma (W_i + b_i)$ 

$$10 * 50 = 500 (hidden layer 1)\\  + 50 * 20 = 1500 (second hidden layer)\\  + 20 * 5 = 1600 (output layer)\\ + 75 = 1675 (bias terms for h1,2 and output)$$
