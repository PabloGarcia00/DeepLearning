# Monday: Convolutional networks

Multilayer perceptrons models the relationship with every other feature. However with a 1 MPixel image we get 1 billion learnable weights. Furthermore, to make the model learn we would need around a dataset that is tantamount to the number of parameters that need to be estimated. 

## How do we simplify the problem 

remove unimportant details:
* location of the object.
* Global context: do we need to see all of the image.

We do this by using convultional filters/kernels that slide over the image. A CNN usually learns a few kernels, its properties are: 
* local: they convolve the values of the image, roll values into one.
* shared: the same filter is applied everywhere.
* many filters are learned in parallel
* the kernel size depends on the number of color bands in the image. 

say we have 6 kernels, than the output is 6 activation maps. 

## padding, stride and pooling 

During convolutions the size of the image changes e.g. 10 rounds of convolution using a 5x5 kernel on a 240x240 image results in a a loss of 40 pixels on both dimensions. Padding adds dead(black) pixels back to the borders to remedy this. 

* padding: adding dead pixels to remedy loss of pixels during convolution
* stride: the pace at which a kernel slides over the image.
* pooling: find the highest, lowest or average value of a matrix/image. Allows for more invariate feature maps.

the new image dimensions can be calculated by the following formula:
$$\frac{image - kernel + 1}{stride}$$
